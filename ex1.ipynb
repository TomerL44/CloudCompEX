```python
from typing import Text
import ipywidgets as widgets



with open('drive/MyDrive/test/students.txt') as file:
  lines = file.readlines()
  lines[0] = lines[0][1:-1]
  students = {}
  for line in lines:
    student = line.split("~")
    students[student[0]] = student[1:]
  file.close()

w = widgets.Dropdown(
    options=list(students.keys()),
    value=list(students.keys())[0],
    description='Student:',
)
mail = widgets.Text()
courses = widgets.Text()
link = widgets.Text()
fav_show = widgets.Text()

fav_show.placeholder = "favorite show"



def on_fav_show_submit(sender):
    with open('drive/MyDrive/test/students.txt','a') as file:
      file.write('\n new favourite show for ' + w.value + ' ' + sender.value)
      file.close()

fav_show.on_submit(on_fav_show_submit)

def on_change(change):
    if change['type'] == 'change' and change['name'] == 'value':
        mail.value = students[change['new']][0]
        courses.value = students[change['new']][1]
        link.value = students[change['new']][2]




w.observe(on_change)

display(fav_show)
display(w)
display(mail)
display(courses)
display(link)


```
