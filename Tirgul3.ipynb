```python
from firebase import firebase

# Initialize Firebase connection
FBconn = firebase.FirebaseApplication('https://tirgul3-c0e7b-default-rtdb.europe-west1.firebasedatabase.app/', None)

# Retrieve all words
def get_all_words():
    return FBconn.get('/words/', None)

# Add or Update count
def add_or_update_word(word, count = 1):
    # Get the word from db if already exists in db
    data = FBconn.get('/words/', word)
    if data:
        # Update the count
        new_count = data['count'] + count
    else:
        new_count = count
    FBconn.put('/words/', word, {'count': new_count})

# Delete word
def delete_word(word):
    return FBconn.delete('/words/', word)

# Display words
def display_words():
    words = get_all_words()
    # db is empty
    if not words:
        print("\nNo words found.")
        return
    print("\nWord Counts:")
    for word, data in words.items():
        print(f"{word}: {data['count']}")

# Action 1
def add_single_word():
    word = input("Enter a word to add: ").lower()
    add_or_update_word(word)
    print("Word added!")

# Action 2
def add_text():
    words = input("Enter text: ").lower().split()
    for word in words:
        add_or_update_word(word)
    print("Text added!")

# Action 3
def update_word_count():
    display_words()
    word = input("Enter a word to update: ").lower()
    count = int(input("Enter the new count: "))
    add_or_update_word(word, count)
    print("Word count updated!")

# Action 4
def delete_word_action():
    display_words()
    word = input("Enter a word to delete: ").lower()
    delete_word(word)
    print("Word deleted!")

# Menu
while True:
    print("\nCommon Words Counter Menu:")
    print("1. Add single word")
    print("2. Add text for analysis")
    print("3. Update word count")
    print("4. Delete word")
    print("5. View all words")
    print("6. Exit")

    choice = input("\nSelect an option (1-6): ")

    if choice == '1':
        add_single_word()

    elif choice == '2':
        add_text()

    elif choice == '3':
        update_word_count()

    elif choice == '4':
        delete_word_action()

    elif choice == '5':
        display_words()

    elif choice == '6':
        print("Exiting...")
        break

    else:
        print("Invalid choice.")
```
