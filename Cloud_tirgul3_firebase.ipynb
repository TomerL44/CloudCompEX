```python
import re
from firebase import firebase

# Initialize Firebase connection
FIREBASE_URL = "https://tirgul2-57c41-default-rtdb.firebaseio.com/"
fb = firebase.FirebaseApplication(FIREBASE_URL, None)

def get_all_counts():
    """Helper to fetch the current word counts from Firebase."""
    result = fb.get('/word_counts', None)
    return result if result else {}

def sync_to_firebase(word, count):
    """Helper to update a specific word count in Firebase."""
    fb.put('/word_counts', word, count)

def delete_from_firebase(word):
    """Helper to delete a word from Firebase."""
    fb.delete('/word_counts', word)

def add_single_word():
    word = input("Enter a single word to add: ").strip().lower()
    if word:
        word_counts = get_all_counts()
        new_count = word_counts.get(word, 0) + 1
        sync_to_firebase(word, new_count)
        print(f"'{word}' added to Firebase. Current count: {new_count}")
    else:
        print("No word entered.")

def add_full_text():
    text = input("Enter full text for analysis: ").lower()
    words = re.findall(r'\b\w+\b', text)
    if words:
        word_counts = get_all_counts()
        for word in words:
            word_counts[word] = word_counts.get(word, 0) + 1
            sync_to_firebase(word, word_counts[word])
        print("Text analyzed and data synced to Firebase.")
    else:
        print("No words found in the entered text.")

def update_word_count():
    word = input("Enter the word to update: ").strip().lower()
    word_counts = get_all_counts()
    if word not in word_counts:
        print(f"'{word}' not found in database.")
        return
    try:
        new_count = int(input(f"Enter the new count for '{word}' (current: {word_counts[word]}): "))
        if new_count >= 0:
            sync_to_firebase(word, new_count)
            print(f"Count for '{word}' updated to {new_count} in Firebase.")
        else:
            print("Count cannot be negative.")
    except ValueError:
        print("Invalid input. Please enter a number.")

def delete_word():
    word = input("Enter the word to delete: ").strip().lower()
    word_counts = get_all_counts()
    if word in word_counts:
        delete_from_firebase(word)
        print(f"'{word}' deleted from Firebase successfully.")
    else:
        print(f"'{word}' not found in database.")

def view_all_words():
    word_counts = get_all_counts()
    if not word_counts:
        print("No words saved in Firebase currently.")
        return
    print("\n--- Current Firebase Word Counts ---")
    for word, count in word_counts.items():
        print(f"{word}: {count}")
    print("---------------------------\n")

def main_menu():
    while True:
        print("\n--- Firebase Word Counter Menu ---")
        print("1. Add single word")
        print("2. Add text for analysis")
        print("3. Update word count")
        print("4. Delete word")
        print("5. View all words")
        print("6. Exit")

        choice = input("Select an option (1-6): ").strip()

        if choice == '1':
            add_single_word()
        elif choice == '2':
            add_full_text()
        elif choice == '3':
            update_word_count()
        elif choice == '4':
            delete_word()
        elif choice == '5':
            view_all_words()
        elif choice == '6':
            print("Exiting. Goodbye!")
            break
        else:
            print("Invalid choice.")

if __name__ == "__main__":
    main_menu()
```
